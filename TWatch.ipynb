```python
# @title üì∫ 1. GERAR V√çDEO YOUTUBE (Hardsub Discreto)
# @markdown Fonte Pequena (18), Branca. Ideal para n√£o cobrir legendas CC.

import os
import subprocess
import glob
import sys

# 1. PREPARA√á√ÉO (Instala√ß√£o e Drive)
if 'google.colab' in sys.modules:
    from google.colab import drive
    if not os.path.exists('/content/drive'):
        drive.mount('/content/drive')
    subprocess.run("apt-get update -qq && apt-get install -y -qq ffmpeg expect", shell=True, check=True)

def detectar_resolucao(video_path):
    try:
        cmd = ['ffprobe', '-v', 'error', '-select_streams', 'v:0', '-show_entries', 'stream=width,height', '-of', 'csv=p=0', video_path]
        res = subprocess.run(cmd, capture_output=True, text=True).stdout.strip().split(',')
        return int(res[0]), int(res[1])
    except: return None, None

def criar_video(base_path="/content/drive/MyDrive/VIDEOS"):
    os.chdir(base_path)
    print(f"üöÄ INICIANDO MODO YOUTUBE (Discreto)...")

    if not os.path.exists("v.mp4"): return print("‚ùå ERRO: v.mp4 n√£o encontrado!")

    # Configura√ß√£o
    logos = glob.glob("Logo*.png")
    logo = logos[0] if logos else None
    srt = "v.srt" if os.path.exists("v.srt") else None

    if not logo and not srt: return print("‚ùå Nada para queimar.")

    # Comando NVIDIA
    cmd = ['ffmpeg', '-hwaccel', 'cuda', '-i', 'v.mp4']
    if logo: cmd.extend(['-i', logo])

    filter_complex = ""
    if srt and logo:
        filter_complex = f"[0:v]format=yuv420p,subtitles={srt}:force_style='Fontsize=18,PrimaryColour=&H00FFFFFF,OutlineColour=&H00000000,Outline=1,Shadow=0',hwupload_cuda[v0];[1:v]hwupload_cuda[v1];[v0][v1]overlay_cuda=x=0:y=0[vout]"
    elif srt:
        filter_complex = f"[0:v]format=yuv420p,subtitles={srt}:force_style='Fontsize=18,PrimaryColour=&H00FFFFFF,OutlineColour=&H00000000,Outline=1,Shadow=0'[vout]"
    elif logo:
        filter_complex = f"[0:v]format=yuv420p,hwupload_cuda[v0];[1:v]hwupload_cuda[v1];[v0][v1]overlay_cuda=x=0:y=0[vout]"

    cmd.extend(['-filter_complex', filter_complex, '-map', '[vout]', '-map', '0:a', '-c:v', 'h264_nvenc', '-qp', '27', '-preset', 'p1', '-c:a', 'copy', 'queimado.mp4', '-y'])

    import shlex
    os.system(f"unbuffer {' '.join(shlex.quote(x) for x in cmd)}")
    print("\n‚úÖ CONCLU√çDO: queimado.mp4")

if __name__ == "__main__": criar_video()
```

```python
# @title üé¨ 4. COLAR CAPA + AJUSTAR LEGENDA
# @markdown Cola a capa de 1s e gera o arquivo .srt sincronizado.

import os
import subprocess
import re
import sys

if 'google.colab' in sys.modules:
    from google.colab import drive
    if not os.path.exists('/content/drive'): drive.mount('/content/drive')
    subprocess.run("apt-get update -qq && apt-get install -y -qq ffmpeg expect", shell=True, check=True)

def ajustar_srt(entrada, saida, segundos=1):
    try:
        with open(entrada, 'r', encoding='utf-8') as f: linhas = f.readlines()
        output = []
        for linha in linhas:
            match = re.match(r'(\d{2}):(\d{2}):(\d{2}),(\d{3})\s*-->\s*(\d{2}):(\d{2}):(\d{2}),(\d{3})', linha)
            if match:
                # L√≥gica de soma de tempo simplificada
                t1 = (int(match.group(1))*3600 + int(match.group(2))*60 + int(match.group(3)))*1000 + int(match.group(4)) + 1000
                t2 = (int(match.group(5))*3600 + int(match.group(6))*60 + int(match.group(7)))*1000 + int(match.group(8)) + 1000
                time1 = f"{t1//3600000:02d}:{(t1%3600000)//60000:02d}:{(t1%60000)//1000:02d},{t1%1000:03d}"
                time2 = f"{t2//3600000:02d}:{(t2%3600000)//60000:02d}:{(t2%60000)//1000:02d},{t2%1000:03d}"
                output.append(f"{time1} --> {time2}\n")
            else: output.append(linha)
        with open(saida, 'w', encoding='utf-8') as f: f.writelines(output)
        return True
    except: return False

def criar_capa(base_path="/content/drive/MyDrive/VIDEOS"):
    os.chdir(base_path)
    print(f"üöÄ INICIANDO MODO CAPA...")

    if not os.path.exists("Capa.jpg"): return print("‚ùå ERRO: Capa.jpg n√£o encontrada!")
    video = "queimado.mp4" if os.path.exists("queimado.mp4") else "v.mp4" if os.path.exists("v.mp4") else None
    if not video: return print("‚ùå ERRO: Nenhum v√≠deo encontrado!")

    # Detecta Resolu√ß√£o e FPS
    try:
        cmd = ['ffprobe', '-v', 'error', '-select_streams', 'v:0', '-show_entries', 'stream=width,height,r_frame_rate', '-of', 'csv=p=0', video]
        res = subprocess.run(cmd, capture_output=True, text=True).stdout.strip().split(',')
        w, h, fps_raw = res[0], res[1], res[2]
    except: w, h, fps_raw = 1920, 1080, "25"

    # Cria Capa
    os.system(f'ffmpeg -loop 1 -i Capa.jpg -f lavfi -i anullsrc=channel_layout=stereo:sample_rate=44100 -c:v libx264 -c:a aac -pix_fmt yuv420p -s {w}x{h} -r {fps_raw} -t 1 capa.mp4 -y')

    # Concatena
    with open("lista.txt", "w") as f: f.write(f"file 'capa.mp4'\nfile '{video}'\n")
    os.system('ffmpeg -f concat -safe 0 -i lista.txt -c copy limpo.mp4 -y')

    # Ajusta Legenda
    if os.path.exists("v.srt"):
        ajustar_srt("v.srt", "v_ajustado.srt")
        os.system('ffmpeg -i "limpo.mp4" -i "v_ajustado.srt" -map 0:v -map 0:a -map 1:s -c:v copy -c:a copy -c:s mov_text "limpo_final.mp4" -y')
        os.replace("limpo_final.mp4", "limpo.mp4")
        os.system('ffmpeg -i "limpo.mp4" -map 0:s:0 -c:s srt "limpo.srt" -y 2>/dev/null')
        print("‚úÖ CONCLU√çDO: limpo.mp4 + limpo.srt")
    else:
        print("‚úÖ CONCLU√çDO: limpo.mp4 (Sem legenda)")

    # Limpeza
    if os.path.exists("capa.mp4"): os.remove("capa.mp4")
    if os.path.exists("lista.txt"): os.remove("lista.txt")
    if os.path.exists("v_ajustado.srt"): os.remove("v_ajustado.srt")

if __name__ == "__main__": criar_capa()
```
